```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
#Filter the data for movies released in the 1990s

## Filter only movies from the 1990s
movies_1990s = netflix_df[
    (netflix_df['type']=='Movie')&
    (netflix_df['release_year'] >= 1990)&
    (netflix_df['release_year'] <= 1999)
     
]

print(movies_1990s.head())
```

```python
#Step 2: Find the most frequent movie duration

# Ensure the 'duration' column is of string type before using .str accessor
movies_1990s['duration'] = movies_1990s['duration'].astype(str)

# Clean the 'duration' column to remove " min" and convert to integer
movies_1990s['duration_clean'] = movies_1990s['duration'].str.replace(' min', '', regex=False).astype(int)

# Find the most frequent duration (mode)
duration = int(movies_1990s['duration_clean'].mode()[0])

print(duration)
```

```python
#Step 3:  A short movie is < 90 minutes and genre includes "Action"





short_action_movies = movies_1990s[
    (movies_1990s['duration_clean'] < 90 )&
    (movies_1990s['genre'].str.contains('Action' ,na=False))
]

short_movie_count = len(short_action_movies)
```

```python
print("Most frequent duration :", duration)
print("Number of short action movies :",short_movie_count)
```
